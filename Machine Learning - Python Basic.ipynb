### Matrix Multiplication

**A [l, m] * B [m, n] = C [l, m].** The first parameter is rows number and second parameter is columns number. 

- **columns number from matrix A has to be the same than rows number from matrix B. The output matrix will have dimensions (l, m)**
- **The command to do it is dot( ).**

```python
a = np.array([[3, 6, 7], [5, -3, 0]])
print(a.shape)
b = np.array([[1, 1], [2, 1], [3, -3]])
print(b.shape)
c = a.dot(b)
print(c)
```

## Scikit Learn

```python
movies2 = pd.read_csv('Notebooks_Platzi_ML/machine-learning-platz-master/datasets/numbers.csv', encoding='utf-8')
movies2
```

### Merge (joins) 

```python
movies2_filtered = movies2[['movie_title','production_budget','worldwide_gross']]
display(movies2_filtered.shape)
# In this dataframe We have 4341 records
display(movies.shape)
# In the original dataframe We have 5043 records
```

**When We merge both dataframes, We will loss 15% of the original data but it is more important to ensure the data quality**

```python
# to concatenate 2 Dataframes:

# By default the concatenate function is line by line (vertically). To do it column by column set up 'axis=1'

newNumericDF = pd.concat([numericDF,stringDF['movie_title']],axis=1).
display(newNumericDF)
```

```python
# Merge the newNumericd DF with movies 2 (new gross data wrappied)

mergedDF = pd.merge(newNumericDF, movies2_filtered, on='movie_title', how='right')
display(mergedDF)
```

### How to handle Missing Data? 

```python
# mergedDF.notnull()   # This funcion returns True if exist data or False if not cell by cell

mergedDF.notnull().apply(pd.Series.value_counts)
```

```python
(mergedDF != 0).apply(pd.Series.value_counts)
```

**Our target variable is 'Worldwide_gross' so, in order to improve the data quality We will remove the data that not exist and is equal to 0**.

```python
availableDF = ((mergedDF != 0) & (mergedDF.notnull())) # Return False if data is 0 or if not exist.
availableDF.all(axis=1).value_counts() # axis 1 allows move for all columns
```

```python
mask = availableDF['worldwide_gross']
mergedDF = mergedDF[mask]
```

```python
((mergedDF != 0) & (mergedDF.notnull())).worldwide_gross.value_counts()
```

### How to complete Missing Data?

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')  
# instance. Missing values will be NaN type and these values will be replaced by the column mean.
```

From mergedDF we need to remove no numeric fields ('movie title') and delete the duplicated column ('duration.1') 

```python
mergedDF = mergedDF.drop(columns=['movie_title','duration.1'])
```

```python
values = imputer.fit_transform(mergedDF)
```

#### Now, We will create the final DF replacing the NaN values. 

```python
X = pd.DataFrame(values)
X.columns = mergedDF.columns
X.index = mergedDF.index
X.head()
```

```python
X.to_csv('Notebooks_Platzi_ML/machine-learning-platz-master/datasets/intermediate_results/X.csv', index=False)
# Index=False to don't print the index names
```
